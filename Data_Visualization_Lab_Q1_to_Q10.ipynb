```python
# Q1) Download the House Pricing dataset from Kaggle and map the values to Aesthetics

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Housing dataset
housing_df = pd.read_csv('/content/Housing.csv')

# Plot a histogram of the price variable
sns.histplot(data=housing_df, x="price")
plt.title("Distribution of House Prices")
plt.xlabel("Price ($)")
plt.ylabel("Frequency")
plt.show()

# Create a scatterplot of the relationship between area and price
sns.scatterplot(data=housing_df, x="area", y="price")
plt.title("Relationship between House Area and Price")
plt.xlabel("Area (sqft)")
plt.ylabel("Price ($)")
plt.show()

# Plot a bar chart of the number of houses by number of bedrooms
sns.countplot(data=housing_df, x="bedrooms")
plt.title("Number of Houses by Number of Bedrooms")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Count")
plt.show()

# Create a stacked bar chart of the number of houses by number of bedrooms and main road availability
sns.countplot(data=housing_df, x="bedrooms", hue="mainroad")
plt.title("Number of Houses by Number of Bedrooms and Main Road Availability")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Count")
plt.legend(title="Main Road")
plt.show()

# Plot a boxplot of the price variable by number of stories
sns.boxplot(data=housing_df, x="stories", y="price")
plt.title("Price by Number of Stories")
plt.xlabel("Number of Stories")
plt.ylabel("Price ($)")
plt.show()

# Create a heatmap of the correlation matrix of the numeric variables
numeric_vars = ["price", "area", "bedrooms", "bathrooms", "stories", "parking"]
corr = housing_df[numeric_vars].corr()
sns.heatmap(corr, cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Correlation Matrix of Numeric Variables")
plt.show()

# Create a stacked bar chart of the number of houses by preferred area and furnishing status
sns.countplot(data=housing_df, x="prefarea", hue="furnishingstatus")
plt.title("Number of Houses by Preferred Area and Furnishing Status")
plt.xlabel("Preferred Area")
plt.ylabel("Count")
plt.legend(title="Furnishing Status")
plt.show()
```

```python
# Q2) Use different Color scales on the Rainfall Prediction dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the district-wise rainfall dataset
rainfall_df = pd.read_csv('/content/district wise rainfall normal.csv')

# Create a scatterplot of the annual rainfall by district
sns.scatterplot(data=rainfall_df, x="DISTRICT", y="ANNUAL", hue="ANNUAL", palette="viridis")
plt.title("Annual Rainfall by District")
plt.xlabel("District")
plt.ylabel("Rainfall (mm)")
plt.xticks(rotation=90)
plt.show()

# Create a heatmap of the rainfall data by district and month
rainfall_pivot = rainfall_df.pivot_table(index="DISTRICT", values=["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"])
sns.heatmap(data=rainfall_pivot, cmap="YlGnBu")
plt.title("Rainfall by District and Month")
plt.xlabel("Month")
plt.ylabel("District")
plt.show()

# Create a stacked bar chart of the rainfall data by district and season
rainfall_df["Jan-Feb"] = rainfall_df["JAN"] + rainfall_df["FEB"]
rainfall_df["Mar-May"] = rainfall_df["MAR"] + rainfall_df["APR"] + rainfall_df["MAY"]
rainfall_df["Jun-Sep"] = rainfall_df["JUN"] + rainfall_df["JUL"] + rainfall_df["AUG"] + rainfall_df["SEP"]
rainfall_df["Oct-Dec"] = rainfall_df["OCT"] + rainfall_df["NOV"] + rainfall_df["DEC"]
seasonal_pivot = rainfall_df.pivot_table(index="DISTRICT", values=["Jan-Feb", "Mar-May", "Jun-Sep", "Oct-Dec"])
seasonal_pivot.plot(kind="bar", stacked=True, colormap="Paired")
plt.title("Rainfall by District and Season")
plt.xlabel("District")
plt.ylabel("Rainfall (mm)")
plt.xticks(rotation=90)
plt.show()
```

```python
#Q4) Show an example of Skewed data and removal of skewedness

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate skewed data
data = np.random.gamma(1, 3, 1000)

# Plot histogram to visualize skewness
plt.hist(data, bins=50)
plt.title("Skewed Data Histogram")
plt.show()

# Calculate skewness of data
skewness = stats.skew(data)
print("Skewness:", skewness)

# Apply log transformation to reduce skewness
log_data = np.log(data)

# Plot histogram of transformed data
plt.hist(log_data, bins=50)
plt.title("Transformed Data Histogram")
plt.show()

# Calculate skewness of transformed data
log_skewness = stats.skew(log_data)
print("Transformed Skewness:", log_skewness)
```

```python
#Q5) For a sales dataset do a Time Series Visualization?

import pandas as pd
import matplotlib.pyplot as plt

# Load sales data from CSV file
sales_df = pd.read_csv('/content/sales.csv')

# Convert the 'orderdate_year' and 'orderdate_month' columns to a datetime object
sales_df['orderdate'] = pd.to_datetime(sales_df['orderdate_year'].astype(str) + '-' + sales_df['orderdate_month'].astype(str))

# Set the 'orderdate' column as the index
sales_df.set_index('orderdate', inplace=True)

# Aggregate sales by year and month
sales_by_month = sales_df['sales'].resample('M').sum()

# Create a line chart of sales by year and month
plt.plot(sales_by_month.index, sales_by_month)
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales by Year and Month')
plt.show()
```

```python
#Q6) Build a Scatterplot and suggest dimension reduction.



import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate random data for the example
np.random.seed(42)
data = {
    'X': np.random.rand(100),
    'Y': np.random.rand(100),
}
df = pd.DataFrame(data)

# Create the scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(df['X'], df['Y'], c='b', marker='o', label='Data Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot of X vs Y')
plt.legend()
plt.grid(True)
plt.show()




# Dimensionality reduction




from sklearn.decomposition import PCA

# Assuming you have the data in the DataFrame 'df'
# Remove any non-numeric columns if necessary
numeric_df = df.select_dtypes(include=[np.number])

# Create a PCA object with the number of desired components (let's say 2)
pca = PCA(n_components=2)

# Fit and transform the data to the new reduced feature space
reduced_data = pca.fit_transform(numeric_df)

# Create a DataFrame to store the reduced data along with any relevant labels
reduced_df = pd.DataFrame(reduced_data, columns=['PC1', 'PC2'])

# Visualize the scatterplot of the reduced data
plt.figure(figsize=(8, 6))
plt.scatter(reduced_df['PC1'], reduced_df['PC2'], c='r', marker='o', label='Reduced Data Points')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.title('Scatterplot of Reduced Data (PCA)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
#Q8) Create the a trend line with a confidence band in any suitable dataset



import numpy as np
import matplotlib.pyplot as plt
from scipy import stats  # Import the stats module from scipy

# Generate random data with an increasing trend
np.random.seed(42)
x = np.arange(1, 101)  # X-axis values from 1 to 100
y = 2 * x + np.random.normal(0, 10, size=len(x))  # Y-axis values with a linear trend and noise

# Calculate the trend line (linear regression)
coefficients = np.polyfit(x, y, 1)
trend_line = np.polyval(coefficients, x)

# Calculate the confidence band
# Assuming 95% confidence, you can adjust the confidence level if needed
confidence = 0.95
errors = y - trend_line
std_error = np.std(errors)
z_score = np.abs(stats.norm.ppf((1 - confidence) / 2))
confidence_band = z_score * std_error

# Plot the data, trend line, and confidence band
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data Points', color='blue', alpha=0.6)
plt.plot(x, trend_line, label='Trend Line', color='orange', linewidth=2)
plt.fill_between(x, trend_line - confidence_band, trend_line + confidence_band, color='orange', alpha=0.3, label='Confidence Band')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Trend Line with Confidence Band')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q9) Illustrate Partial Transparency and Jittering

#Partial Transparency

import numpy as np
import matplotlib.pyplot as plt

# Generate random data with two classes
np.random.seed(42)
x_class1 = np.random.normal(10, 2, 100)
y_class1 = np.random.normal(8, 2, 100)

x_class2 = np.random.normal(14, 2, 100)
y_class2 = np.random.normal(12, 2, 100)

# Create a scatter plot with partial transparency
plt.figure(figsize=(8, 6))
plt.scatter(x_class1, y_class1, c='b', label='Class 1', alpha=0.6)
plt.scatter(x_class2, y_class2, c='r', label='Class 2', alpha=0.6)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Partial Transparency Scatter Plot')
plt.legend()
plt.grid(True)
plt.show()


#Jittering


import numpy as np
import matplotlib.pyplot as plt

# Generate random data with a linear relationship
np.random.seed(42)
x = np.arange(1, 101)
y = 2 * x + np.random.normal(0, 10, size=len(x))

# Add jitter to the x-axis
jitter = 2
x_jittered = x + np.random.uniform(-jitter, jitter, size=len(x))

# Create a scatter plot with jittering
plt.figure(figsize=(10, 6))
plt.scatter(x_jittered, y, c='b', marker='o', label='Data Points', alpha=0.7)
plt.plot(x, 2 * x, c='r', label='True Trend Line', linewidth=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Jittering')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q10)	Illustrate usage of different color codes



import matplotlib.pyplot as plt

# Sample data for the bar chart
categories = ['Category 1', 'Category 2', 'Category 3']
values = [25, 40, 30]

# Using RGB color code (decimal values in the range 0-1)
rgb_color = (0.2, 0.4, 0.6)

# Using HEX color code (starts with '#')
hex_color = '#FF5733'

# Using named color from Matplotlib's built-in colors
named_color = 'mediumseagreen'

# Create the bar chart
plt.figure(figsize=(8, 6))

# Bar with RGB color code
plt.bar(categories, values, color=rgb_color, label='RGB Color')

# Bar with HEX color code
plt.bar(categories, values, color=hex_color, alpha=0.6, label='HEX Color')

# Bar with named color
plt.bar(categories, values, color=named_color, alpha=0.8, label='Named Color')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Different Color Codes in Bar Chart')
plt.legend()
plt.grid(True)
plt.show()
```
